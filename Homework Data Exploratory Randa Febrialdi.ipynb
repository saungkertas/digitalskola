# Data Covid-19 pada Tahun 2022 di Indonesia.

```python
# Import library
import pandas as pd
import datetime as dt
import seaborn as sns
```

```python
# Read data
df = pd.read_csv("owid-covid-data.csv", sep = ',')
```

```python
# Output df
df.head()
```

```python
# Pilih location = Indonesia
df = df[df['location'] == 'Indonesia']
df.head() # Menampilkan 5 data teratas
```

```python
# Untuk mereset index supaya mulai dari 0 lagi
df = df.reset_index(drop=True)
df.head() 
```

```python
# Melihat informasi data
df.info()
```

```python
# Memilih kolom yg diinginkan (filter kolom)
df = df.filter(['location', 'date', 'total_cases', 'new_cases', 'total_deaths', 'new_deaths'])
df.head()
```

```python
# Hapus kolom NaN
df = df.dropna(axis = 0)
df.head()
```

```python
# Memilih data tahun 2022 saja
df = df[df['date'] >= '2022-01-01'].reset_index(drop=True)
df.head()
```

```python
# Mengubah tipe data
df = df.astype({'total_cases' : int, 'new_cases' : int, 'total_deaths' : int, 'new_deaths' : int})
```

```python
# Mengubah tipe data kolom date dari string menjadi datetime
df['date']= pd.to_datetime(df['date'])
```

```python
# Melihat info data (tipe data apakah sudah berubah)
df.info()
```
